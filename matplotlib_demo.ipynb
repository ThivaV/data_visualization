# Matplotlib Tutorial

```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 25)
y = x * x + 2
print(x)
print(y)
print(np.array([x, y]).reshape(25, 2).reshape(2, 25))
```

```python
# 'r' for RED
pylab.plot(x, y, 'r')
```

```python
# drawing subgraph
pylab.subplot(1, 2, 1) # the contents of the brackets represents (rows, columns, indexes)
pylab.plot(x, y, 'r--') # the thrid parameters here determines colour and line style
pylab.subplot(1, 2, 2)
pylab.plot(y, x, 'g*-')
```
